```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
#You have to turn second order ODE into system of ODEs
#x''=-x'^2 + sin(x) and you need to say x'=v is a new ODE where v is a new variable
# x'=v and v'=-v^2 + sin(x)
def dSdx(x, S):
    x, v= S #order
    return[v,
           -v**2 + np.sin(x)]
x_0= 0    
v_0= 5
S_0= (x_0,v_0)
```

```python
t= np.linspace(0,1,100)
sol= odeint(dSdx, y0=S_0, t=t, tfirst=True) 
```

```python
x_sol = sol.T[0] #x
v_sol = sol.T[1] #x'
```

```python
plt.plot(t, x_sol)
plt.plot(t, v_sol) 
```
