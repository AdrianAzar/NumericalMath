```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
#system of ODEs
def dSdx(x,S): #S=[y1,y2] and dS/dx=[y1',y2']=[... , ...]
    y1, y2= S
    return [y1 + y2**2 + 3*x, 3*y1 + y2**3 -np.cos(x)] #y1'=y1+y2^2+3x and y2'=3y1+y2^3 -cos(x)
y1_0 = 0
y2_0 = 0
S_0 = (y1_0 , y2_0) #vector of initial condition
```

```python
x= np.linspace(0,1,100)
sol= odeint(dSdx, y0=S_0 , t=x , tfirst=True) #t(times)=x times
```

```python
y1_sol= sol.T[0]
y2_sol= sol.T[1]
```

```python
plt.plot(x, y1_sol)
plt.plot(x, y2_sol)
```
