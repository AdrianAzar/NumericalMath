```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
#x2'''=-x1''^3 +x2'+x1+sin(t) and x1'''=-2x2'^2 +x2
#I defined the following: v1=x1' v2=x2' a1=x1'' a2=x2''
#Now: a2'=-a1^3 + v2 +sin(t) and a1'=-2v2^2 + x2
def dSdt(t,S):
    x1,v1,a1,x2,v2,a2= S
    return[v1,
           a1,
           -2*v2**2 + x2,
           v2,
           a2,
           -a1**3 + v2 + np.sin(t)]
x1_0=0
v1_0=0
a1_0=0
x2_0=0
v2_0=0
a2_0=0
S_0=(x1_0, v1_0, a1_0, x1_0, v1_0, a1_0)
```

```python
t =np.linspace(0,1,100)
sol = odeint(dSdt, y0=S_0, t=t, tfirst=True)
```

```python
plt.plot(t,sol.T[0])
```
