```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
def dvdt(t,v):       #v'=av^2-b
    return 3*v**2 -5
v0=0
```

```python
t=np.linspace(0,1,100)
sol_m1= odeint(dvdt, y0=v0, t=t, tfirst=True) #using odeint #y0 is initial cond 
sol_m2= solve_ivp(dvdt, t_span=(0,max(t)), y0=[v0], t_eval=t) #using solve_ivp #t_span to known the range #y0 has an array because of systems of equ
```

```python
sol_m1.T[0] #transpose for clean solution and for index 0(only 1 ode) 
```

```python
 sol_m2.y #.y for the solution
```

```python
v_sol_m1= sol_m1.T[0]
v_sol_m2= sol_m2.y[0]
```

```python
plt.plot(t,v_sol_m1)
plt.plot(t,v_sol_m2,'--')
plt.xlabel('t')
plt.ylabel('v(t)')
plt.show()
```
