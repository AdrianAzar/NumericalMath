```python
import numpy as np
import matplotlib.pyplot as plt
def f(x,y):
    return x*y


a=float(input('a= '))
b=float(input('b= '))
N=int(input('max= '))
h=(b-a)/N
x=np.zeros(N+1)
y=np.zeros(N+1)
y[0]=1
x[0]=a
for i in range(N):
    k1=h*f(x[i],y[i])
    k2=h*f(x[i]+0.5*h,y[i]+0.5*k1)
    k3=h*f(x[i]+0.5*h,y[i]+0.5*k2)
    k4=h*f(x[i]+h,y[i]+k3)
    y[i+1]=y[i]+1/6*(k1+2*(k2+k3)+k4)
    x[i+1]=x[0]+i*h

plt.plot(x,y)
plt.show()   
```
